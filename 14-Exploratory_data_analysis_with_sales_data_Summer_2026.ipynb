### Question 1: Import Data and Store-wise Revenue Analysis (4 Marks)

Import `pandas` and `numpy`. Download the dataset `sales.csv` from https://drive.google.com/drive/folders/1gieHICVDBbUKMZiSF4YRQMAJic-w50JM?usp=sharing. Load the dataset `sales.csv` into a DataFrame. Then, find the top 5 stores with the highest total sales and the bottom 5 stores with the lowest total sales.

**Tasks:**
- Aggregate total sales by store_id
- Rank stores by total revenue
- Compare average sales per department within those stores

```python
# Import required libraries
import pandas as pd # to make data in table format
import numpy as np  # to handle those data
import matplotlib.pyplot as plt # to plot those data & lines for visuilazation

# Load the dataset sales.csv into a DataFrame
# (Assuming the file is downloaded locally or available in your working folder)
df = pd.read_csv("sales.csv")

# Task 1: Aggregate total sales by store_id
store_revenue = df.groupby("store_id")["weekly_sales"].sum().reset_index()

# Task 2: Rank stores by total revenue
store_revenue_sorted = store_revenue.sort_values(by="weekly_sales", ascending=False).reset_index(drop=True)

print("--- Top 5 Stores with Highest Total Sales ---")
top_5_stores = store_revenue_sorted.head(5)
print(top_5_stores)

print("\n--- Bottom 5 Stores with Lowest Total Sales ---")
bottom_5_stores = store_revenue_sorted.tail(5)
print(bottom_5_stores)

# Task 3: Compare average sales per department within those top and bottom stores
top_5_ids = top_5_stores["store_id"].tolist()
bottom_5_ids = bottom_5_stores["store_id"].tolist()

print("\n--- Top 5 Stores: Average Sales per Department ---")
top_dept_avg = df[df["store_id"].isin(top_5_ids)].groupby(["store_id", "department"])["weekly_sales"].mean().reset_index()
print(top_dept_avg.head(10)) # showing first 10 rows for display

print("\n--- Bottom 5 Stores: Average Sales per Department ---")
bottom_dept_avg = df[df["store_id"].isin(bottom_5_ids)].groupby(["store_id", "department"])["weekly_sales"].mean().reset_index()
print(bottom_dept_avg.head(10)) # showing first 10 rows for display
```

### Question 2: Department Performance Consistency (3 Marks)

Which departments have the most stable weekly sales and which are the most volatile?

**Tasks:**
- Use standard deviation and coefficient of variation
- Identify highly stable and highly volatile departments

```python
# Task 1: Use standard deviation and coefficient of variation to analyze consistency
# Group by department and calculate mean and standard deviation
dept_stats = df.groupby("department")["weekly_sales"].agg(["mean", "std"]).reset_index()

# Calculate Coefficient of Variation (CV = std / mean)
dept_stats["coefficient_of_variation"] = dept_stats["std"] / dept_stats["mean"]

# Task 2: Identify highly stable and highly volatile departments
# Sorting by CV: Low CV means stable, High CV means volatile
dept_stats_sorted = dept_stats.sort_values(by="coefficient_of_variation").reset_index(drop=True)

print("--- Top 5 Most Stable Departments (Lowest CV) ---")
print(dept_stats_sorted.head(5))

print("\n--- Top 5 Most Volatile Departments (Highest CV) ---")
print(dept_stats_sorted.tail(5))
```

### Question 3: Holiday vs Non-Holiday Sales (3 Marks)

Analyze whether holidays significantly affect weekly sales.

**Tasks:**
- Compare average sales during holidays vs non-holidays
- Calculate percentage increase/decrease
- Identify departments benefiting most from holidays

```python
# Task 1: Compare average sales during holidays vs non-holidays
holiday_comparison = df.groupby("is_holiday")["weekly_sales"].mean().reset_index()
print("--- Average Sales: Non-Holiday (0) vs Holiday (1) ---")
print(holiday_comparison)

# Task 2: Calculate percentage increase/decrease
non_holiday_avg = holiday_comparison.loc[holiday_comparison["is_holiday"] == 0, "weekly_sales"].values[0]
holiday_avg = holiday_comparison.loc[holiday_comparison["is_holiday"] == 1, "weekly_sales"].values[0]
pct_change = ((holiday_avg - non_holiday_avg) / non_holiday_avg) * 100
print(f"\nPercentage increase in sales during holidays: {pct_change:.2f}%")

# Task 3: Identify departments benefiting most from holidays
dept_holiday = df.groupby(["department", "is_holiday"])["weekly_sales"].mean().unstack().reset_index()
# Rename columns for clarity
dept_holiday.columns = ["department", "Non_Holiday_Avg", "Holiday_Avg"]
# Calculate pure dollar increase for each department during holidays
dept_holiday["sales_lift"] = dept_holiday["Holiday_Avg"] - dept_holiday["Non_Holiday_Avg"]

dept_benefiting = dept_holiday.sort_values(by="sales_lift", ascending=False).reset_index(drop=True)
print("\n--- Top 5 Departments Benefiting Most From Holidays ---")
print(dept_benefiting.head(5))
```

### Question 4: Seasonal Trend Detection (3 Marks)

Identify whether sales increase or decrease during specific months.

**Tasks:**
- Extract month from the date column
- Compute average monthly sales
- Determine highest and lowest sales months

```python
# Task 1: Extract month from the date column
df["date"] = pd.to_datetime(df["date"]) # ensuring standard datetime object format
df["month"] = df["date"].dt.month

# Task 2: Compute average monthly sales
monthly_sales = df.groupby("month")["weekly_sales"].mean().reset_index()

# Task 3: Determine highest and lowest sales months
monthly_sales_sorted = monthly_sales.sort_values(by="weekly_sales", ascending=False).reset_index(drop=True)

print("--- Monthly Performance Ranked ---")
print(monthly_sales_sorted)

highest_month = monthly_sales_sorted.iloc[0]
lowest_month = monthly_sales_sorted.iloc[-1]

print(f"\nHighest sales month is Month {int(highest_month['month'])} with average sales of {highest_month['weekly_sales']:.2f}")
print(f"Lowest sales month is Month {int(lowest_month['month'])} with average sales of {lowest_month['weekly_sales']:.2f}")
```

### Question 5: Store and Department Contribution (4 Marks)

Which combinations of store and department contribute the most to overall revenue?

**Tasks:**
- Group using store_id and department
- Find top and worst-performing combinations
- Compute percentage contribution

```python
# Task 1: Group using store_id and department
comb_sales = df.groupby(["store_id", "department"])["weekly_sales"].sum().reset_index()

# Task 3: Compute percentage contribution to overall total corporate revenue
total_corporate_revenue = df["weekly_sales"].sum()
comb_sales["percentage_contribution"] = (comb_sales["weekly_sales"] / total_corporate_revenue) * 100

# Task 2: Find top and worst-performing combinations
comb_sorted = comb_sales.sort_values(by="weekly_sales", ascending=False).reset_index(drop=True)

print("--- Top 5 Store-Department Combinations by Total Revenue ---")
print(comb_sorted.head(5))

print("\n--- Worst 5 Store-Department Combinations by Total Revenue ---")
print(comb_sorted.tail(5))
```

### Question 6: Load MNIST Data (3 Marks)

Load the MNIST dataset using `sklearn.datasets.load_digits`. Separate the dataset into features (`X`) and target labels (`y`).
Print the shape of the features and the target arrays.

**Expected Output:** The shape of `X` and `y`.

```python
# Import MNIST loading mechanism from scikit-learn
from sklearn.datasets import load_digits

# Load the dataset
mnist = load_digits()

# Separate the dataset into features (X) and target labels (y)
X = mnist.data
y = mnist.target

# Print the shape of the features and target arrays
print(f"Shape of features array X: {X.shape}")
print(f"Shape of target array y: {y.shape}")
```

### Question 7: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`). 
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
# Import KMeans estimator from scikit-learn library
from sklearn.cluster import KMeans

# Initialize K-Means clustering model with 10 clusters (digits 0-9)
kmeans = KMeans(n_clusters=10, random_state=42)

# Fit the model on the MNIST features (X)
kmeans.fit(X)

# Assign the cluster labels to a variable
kmeans_labels = kmeans.labels_

print("K-Means Model successfully fitted!")
print("Cluster labels array extracted:")
print(kmeans_labels[:50]) # Printing first 50 labels as an indicator sample
```

### Question 8: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster. 
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
# Import target evaluation metrics from scikit-learn
from sklearn.metrics import f1_score
from scipy.stats import mode # to handle finding the most frequent labels

# Map each arbitrary cluster label to the most frequent true label in that cluster
mapped_labels = np.zeros_like(kmeans_labels)

for i in range(10):
    # Find positions where cluster equals i
    mask = (kmeans_labels == i)
    # Extract the true labels for these positions and find the mode
    if np.sum(mask) > 0:
        true_mode = mode(y[mask], keepdims=True).mode[0]
        mapped_labels[mask] = true_mode

# Calculate and print the macro-averaged F1 score
macro_f1 = f1_score(y, mapped_labels, average='macro')
print(f"Calculated Cluster-Mapped Macro-Averaged F1 Score: {macro_f1:.4f}")
```
